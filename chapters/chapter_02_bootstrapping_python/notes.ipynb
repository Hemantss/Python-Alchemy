## 2.1 Interactive Prompt

Python offers an interactive prompt which can be easyly accessed through:
1. By typing python on OS commad prompt, which will bring up python prompt.
2. By opening Python shell comes with Python installation.
3. Or By opening Python IDLE comes with Python installation.

#### Interactive First Program

```python
name = input("What’s your name?")
print(f"Hello, {name}! Welcome to Python Alchemy.")
```

#### play a small guessing game

```python
import random
print("I’m thinking of a number 1-5...")
guess = int(input("Your guess: "))
print("Correct!" if guess == random.randint(1,5) else "Try again!")
```
